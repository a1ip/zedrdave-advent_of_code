```python
# %%timeit

code = [(l[:3], [int(i) for i in l[4:].split()]) for l in open('08/input.txt')]

def run(prog, IP = 0):
    acc = 0
    steps = []
    
    while 0 <= IP < len(prog):
        steps += [IP]
        cmd, (p1,) = prog[IP]
        
        if cmd == 'jmp':
            IP += p1
        elif cmd == 'acc':
            acc += p1
            IP += 1
        elif cmd == 'nop':
            IP += 1
        
        if IP in steps:
            return False, acc

    return IP == len(prog), acc

print('Part 1:', run(code)[1])

for l,(𝒾,a) in enumerate(code):
    if 𝒾 == 'jmp':
        𝒾 = 'nop'
    elif 𝒾 == 'nop':
        𝒾 = 'jmp'
    b, 𝒜 = run(code[:l] + [(𝒾,a)] + code[l+1:])
    if b:
        break

print('Part 2:', 𝒜)
```

```python
# %%timeit

import networkx as nx

𝓟 = [(l[:3], int(l[4:])) for l in open('08/input.txt')]

G = nx.DiGraph([n, n+(a if c=='jmp' else 1), {'acc': (a if c=='acc' else 0)}]
               for n,(c,a) in enumerate(𝓟))

cy = nx.find_cycle(G, 0)
print('Part 1:', nx.shortest_path_length(G, 0, cy[-1][0], 'acc'))

nx.set_node_attributes(G, {n:{'i':a} for n,a in enumerate(𝓟)})
for s,t in cy:
    c, a = G.nodes[s]['i']
    nt = s + {'jmp': 1, 'nop': a, 'acc': 0}[c]
    if nx.has_path(G, nt, len(𝓟)):
        G.remove_edge(s, t)
        G.add_edge(s, nt, acc=0)
        break
print('Part 2:', nx.shortest_path_length(G, 0, len(𝓟), 'acc'))
```
